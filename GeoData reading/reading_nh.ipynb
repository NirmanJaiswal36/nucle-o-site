```python
def nearest(row, other_gdf, point_column='geometry'):
    # Find the nearest point in the other_gdf's unary union
    nearest_geom = nearest_points(row[point_column], other_gdf.unary_union)[1]
    
    # Find the nearest row by calculating the minimum distance to the nearest point
    min_distance = float('inf')
    nearest_row = None
    
    for idx, highway in other_gdf.iterrows():
        distance = highway.geometry.distance(nearest_geom)
        if distance < min_distance:
            min_distance = distance
            nearest_row = highway
    
    if nearest_row is not None:
        highway_name = nearest_row['name']  # Ensure 'ref' is the column name for highway names
        distance = row[point_column].distance(nearest_row.geometry)
        return nearest_row.geometry, distance, highway_name
    else:
        return None, None, None
```
