```python

base_directory = r"D:\CS\GitHub\Nuclear-Site-Selection\state_wise_grid_points"
def create_directory(state_name):
    
    safe_state_name = state_name.replace(' ','_').replace('/','_')
    folder_path = os.path.join(base_directory,safe_state_name)

    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Created folder: {folder_path}")
    else:
        print(f"Folder already exists: {folder_path}")
```
