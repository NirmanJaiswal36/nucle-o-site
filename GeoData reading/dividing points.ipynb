```python
base_dir = r'D:\CS\GitHub\Nuclear-Site-Selection\gridpoints_csv_dams'
dam_name = 'Ajan Dam'
gdf = grid[grid['dam_id']=='Ajan Dam']
def create_directory(dam_name):
    # Create a safe folder name (remove special characters or spaces if needed)
    safe_dam_name = dam_name.replace(' ','_').replace('/','_').replace(':','').replace('\n','_').replace("?","")
    
    # Define the path for the new folder
    folder_path = os.path.join(base_dir, safe_dam_name)
    
    # Create the directory if it does not exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Created folder: {folder_path}")
    else:
        print(f"Folder already exists: {folder_path}")
```

```python
dams['dm_name'].apply(lambda dam_name:create_directory(dam_name))
```
