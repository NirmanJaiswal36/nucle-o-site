```python
roads = gpd.read_file(r"D:\Downloads\western-zone-latest-free.shp\gis_osm_roads_free_1.shp")
```

```python
national_highways = roads[roads['ref'].str.contains('NH', na=False)]
```

```python
shapefile_path = r'D:\Downloads\Administrative Boundary Database\STATE_BOUNDARY.shp' 
gdf = gpd.read_file(shapefile_path)
```

```python
national_highways = national_highways.to_crs(gdf.crs)
```

```python
fig,ax = plt.subplots()
gdf.boundary.plot(ax=ax, color = 'black')
national_highways.plot(ax=ax)
```
